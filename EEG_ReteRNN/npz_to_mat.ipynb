# Funzione npz_to_mat

```python
def npz_to_mat (filename):
  data = np.load('/content/drive/MyDrive/eeg_DNN-Conti/eeg2/' + filename + '.npz')
  newfile= filename+'.mat'
  sio.savemat(newfile, mdict={'eeg':data['eeg']})
  mat=sio.loadmat(newfile)
  return data, mat

  # variabili: data=NpzFile, mat=dict (dizionario)
```

```python
data1,mat1= npz_to_mat ('S01_AD')
data2,mat2= npz_to_mat ('S02_AD')
data3,mat3= npz_to_mat ('S03_AD')
```
